```python
from datasets import load_dataset
wikiann_da = load_dataset("wikiann", "da")
```

```python
wikiann_no = load_dataset("wikiann", "no")
```

```python
# Overview of both DatasetDict objects
print("wikiann_da:")
print(wikiann_da)

print("\nwikiann_no:")
print(wikiann_no)

# Show one sample from each split for each dataset
for name, ds in [("wikiann_da", wikiann_da), ("wikiann_no", wikiann_no)]:
    print(f"\n{name} samples:")
    for split in ["train", "validation", "test"]:
        print(f"\n{split} sample:")
        print(ds[split][0])
```
